```python
def fibonacci(num):
    n1 = 0
    n2 = 1
    count =0
    
    while count<num+1:
        point = n1
        nth = n1+n2
        n1=n2
        n2 = nth
        count+=1
    print(point)
    return 


if __name__ == "__main__":
    N = int(input())
    fibonacci(N)
```

# average of given problems

```python
arr = list(map(int,input().split(" ")))

def sumaverage(arr):
    sumint = 0
    i =0
    while i<len(arr):
        sumint += arr[i]
        i+=1
    return sumint
    
def average(arr):
    avg = sumaverage(arr)/len(arr)
    print(round(avg,2))

average(arr)
```
